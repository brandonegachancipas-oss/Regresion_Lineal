La curva ROC (Receiver Operating Characteristic) es una gráfica que muestra el rendimiento de un modelo de clasificación binaria, se utiliza para elegir el mejor umbral de decisión como detectar más positivos (alta sensibilidad) aunque te equivoques con falsos positivos O evitar falsos positivos aunque se escapen algunos casos.

**Respuesta 2:**
La accuracy mide el porcentaje de predicciones correctas que hace un modelo de clasificación, comparado con el total de predicciones, cuántas veces el modelo acertó (ya sea prediciendo positivos o negativos).
El problema aparece cuando las clases están desbalanceadas (cuando una clase es mucho más frecuente que la otra).

**Respuesta 3:**
La matriz de confusión es una tabla que resume el rendimiento de un modelo de clasificación, mostrando cuántas predicciones fueron correctas y cuántas se equivocaron, diferenciando entre tipos de acierto y error.
Se construye comparando las predicciones del modelo contra los valores reales.

**Componentes:**

*Verdaderos Positivos (TP):
Casos que eran realmente positivos y el modelo los predijo como positivos.

*Falsos Positivos (FP):
Casos que eran realmente negativos, pero el modelo los predijo como positivos.

*Verdaderos Negativos (TN):
Casos que eran realmente negativos y el modelo también predijo negativo.

*Falsos Negativos (FN):
Casos que eran realmente positivos, pero el modelo los predijo como negativos.
